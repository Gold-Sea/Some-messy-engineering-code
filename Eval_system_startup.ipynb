```python
"""
first layer inference start time
"""

%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python

file_name = "Eval_system_startup_v100.pdf"


models = ["ResNet152", "Inception_v3", "Bert_base"]
ready_model = [5.407938957, 9.243407249, 0.9592199326]
our_sys = [9.020881653, 14.06215191, 4.578995705]
# simple_trans = [188.6979866, 106.8055391, 191.4993525]
mps = [20.66297531, 42.87097454, 2.949023247]
kill_restart = [4501.38351, 5527.307458, 6033.676591]

x = np.arange(len(models))
width = 0.8
n = 4

# setup font size
font_size = 16
plt.rc('font',**{'size': font_size, 'family': 'Arial' })

# fig, ax = plt.subplots()
fig, (ax, ax2) = plt.subplots(2, 1, sharex=True, figsize=(8, 3.69),
                              gridspec_kw={'height_ratios': [2, 1.5]})
plt.subplots_adjust(hspace=0.15)

# upper part
rects1 = ax.bar(x - width/n, ready_model, width/n, label='Ready Model', 
                edgecolor='black', linewidth=0.5,
                )
rects2 = ax.bar(x + 2*width/n, kill_restart, width/n, label='Two Processes', 
                edgecolor='black', linewidth=0.5,)
rects3 = ax.bar(x + width/n, mps, width/n, label='MPS', 
                edgecolor='black', linewidth=0.5,)

rects4 = ax.bar(x , our_sys, width/n, label="NeuroLambda",
                linewidth=0.5,
                edgecolor='black')

# bottom part
rects2_1 = ax2.bar(x - width/n, ready_model, width/n, label='Ready Model', 
                edgecolor='black', linewidth=0.5,
                )
rects2_2 = ax2.bar(x + 2*width/n, kill_restart, width/n, label='Two Processes', 
                edgecolor='black', linewidth=0.5,)
rects2_3 = ax2.bar(x + width/n, mps, width/n, label='MPS', 
                linewidth=0.5,
                edgecolor='black')

rects2_4 = ax2.bar(x , our_sys, width/n, label="NeuroLambda",
                linewidth=0.5,
                edgecolor='black')

ax.set_ylim(4000, 6300)
ax2.set_ylim(0, 100)

# hide the spines between ax and ax2
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

ax.tick_params(bottom=False)  # don't put tick labels at the top
ax2.xaxis.tick_bottom()

d = .015  # how big to make the diagonal lines in axes coordinates
# arguments to pass to plot, just so we don't keep repeating them
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
# ax.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
# ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal

ax.yaxis.set_label_coords(-0.1, -0.05)
ax.set_ylabel('Latency (ms)')
# ax.set_title('End to End Latency: vs Simple Transmission (V100)')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend((rects1, rects2, rects3), ('Ready Model', 'Two Processes', 'MPS'), 
        frameon=False, loc='upper left', ncol=1,
          bbox_to_anchor=(0.0, 1.05),
          prop={'size': font_size-1})

ax2.legend(rects4, ["NeuroLambda"],
           loc='upper left',
           bbox_to_anchor=(0.2, 2.05, 0.5, 0.5),
          frameon=False,
          prop={'size': font_size-1})


# fig.tight_layout()
bar_size = 18
ax.text(0.19, 3890, "/",rotation=-55, fontdict={'family': 'serif',
                                    'color':  'black',
                                    'weight': 'normal',
                                    'size': bar_size,
                                    })
ax2.text(0.19, 95, "/",rotation=-55, fontdict={'family': 'serif',
                                    'color':  'black',
                                    'weight': 'normal',
                                    'size': bar_size,
                                    })
ax.text(1.19, 3890, "/",rotation=-55, fontdict={'family': 'serif',
                                    'color':  'black',
                                    'weight': 'normal',
                                    'size': bar_size,
                                    })
ax2.text(1.19, 95, "/",rotation=-55, fontdict={'family': 'serif',
                                    'color':  'black',
                                    'weight': 'normal',
                                    'size': bar_size,
                                    })

ax.text(2.19, 3890, "/",rotation=-55, fontdict={'family': 'serif',
                                    'color':  'black',
                                    'weight': 'normal',
                                    'size': bar_size,
                                    })
ax2.text(2.19, 95, "/",rotation=-55, fontdict={'family': 'serif',
                                    'color':  'black',
                                    'weight': 'normal',
                                    'size': bar_size,
                                    })

# plt.show()
plt.savefig(file_name, format="pdf")
```
